```python
import pandas as pd
```

```python
df = pd.read_csv("raw.csv")
df['ID'] = df['CaseID'].astype(str) + df['contest_no'].astype(str)
df.head(10)
```

```python
profiles = df.iloc[:, 3:12]
profiles.index += 1
profiles.to_csv('profiles.tsv', sep='\t', index_label='Profile')
profiles.tail()
```

```python
comparisons = pd.DataFrame({'Profile1': [], 'Profile2': []})

for id in df['ID'].unique():
    df_id = df[df['ID'] == id]
    row_val = dict()
    row_val['Profile1'] = df_id[df_id['Chosen_Immigrant'] == 1]['Unnamed: 0'].values[0]
    row_val['Profile2'] = df_id[df_id['Chosen_Immigrant'] == 0]['Unnamed: 0'].values[0]
    comparisons = comparisons.append(row_val, ignore_index=True)
    
comparisons.to_csv('comparisons.tsv', sep='\t', index_label='Comparisons')
comparisons.head()
```

```python
n_comparisons = int(comparisons.shape[0])
n_individuals = int(n_comparisons / 5)
selections = pd.DataFrame()

for i in range(n_individuals):
    individual_selections = [0] * n_comparisons
    for j in range(5):
        individual_selections[(5*i)+j] = 1
    df_individual = pd.DataFrame({'Individual{}'.format(i+1): individual_selections})
    selections = pd.concat([selections, df_individual], ignore_index=True, axis=1)
    
selections.columns = column_names = ['Individual{}'.format(i+1) for i in range(n_individuals)]
selections.to_csv('selections.tsv', sep='\t', index_label='Comparisons')
selections.head()
```
